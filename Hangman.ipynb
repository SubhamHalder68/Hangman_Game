### Step 3:Word is take from txt file and call Hangman(main Function)

```python
print('Welcome to Hangman')
print("-------------------")
Play=True
while Play:
    print("try to guess the word in less than 10 attempts")
    #text file where store words
    f=open("hangman.txt",'r')
    #words are taken from file and make as a list
    word_list=f.read().splitlines()
    #word is randomly chosen
    word=random.choice(word_list)
    #hangman call
    hangman(word)
    #replay or not
    while True:
        replay=input('Are you want to replay? if yes then press y and no then press n=>')
        if replay.lower()=='y':
            break
        elif replay.lower()=='n':
            Play=False
            break
        else:
            print('please enter correct input')
```
